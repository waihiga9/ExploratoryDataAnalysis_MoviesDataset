# Genre vs Rating

First, determine the top 10 most popular rated movie genres and give their respective average rating

```python
q = """
SELECT movie_id,
       primary_title,
       genres,
       averagerating,
       numvotes
FROM movie_basics
JOIN movie_ratings
    USING (movie_id)
"""
genre_ratings = pd.read_sql(q, conn)
genre_ratings.shape
```

```python
q = """
SELECT genres,
       COUNT(genres) AS n_genres,
       AVG(averagerating) AS average_rating,
       SUM(numvotes) AS total_votes
FROM movie_ratings
JOIN movie_basics
    USING (movie_id)
GROUP BY genres
ORDER BY n_genres DESC, average_rating
LIMIT 10
"""
pd.read_sql(q, conn)
```

```python
# Top 5 genres
genre_ratings.genres.value_counts().head()
```

```python
sorted(genre_ratings.numvotes, reverse=True)
```

```python
q = """
SELECT movie_id,
       primary_title,
       genres,
       averagerating,
       numvotes
FROM movie_ratings
JOIN movie_basics
    USING (movie_id)
WHERE genres = 'Drama'
"""
pd.read_sql(q, conn).averagerating.describe()
```
